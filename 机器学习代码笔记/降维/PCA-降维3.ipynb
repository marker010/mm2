```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python

# 加载iris数据集
iris = load_iris()
X = iris.data
y = iris.target
```

```python

# PCA降维到3维
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
X_pca
```

```python

# 绘制三维散点图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis')

# 添加图例
legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)

ax.view_init(elev=0, azim=90)  # 你可以调整 elev 和 azim 的值

ax.set_xlabel(iris.target_names[0])
ax.set_ylabel(iris.target_names[1])
ax.set_zlabel(iris.target_names[2])
plt.show()
```
