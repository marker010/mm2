```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
seed = np.random.RandomState(42)
```

```python
X=np.sort(10.5*seed.rand(100,1),axis=0)
print(X)
```

```python
Y=np.sin(X).ravel()
print(Y)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
Y[::5] += 2*(0.5 - seed.rand(20))
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
reg1 = DecisionTreeRegressor(max_depth=2)
reg2 = DecisionTreeRegressor(max_depth=5)
reg1.fit(X,Y)
reg2.fit(X,Y)
```

```python
X_test = np.arange(0,10,0.05)[:,np.newaxis]
X_test
```

```python
Yt1 = reg1.predict(X_test)
Yt2 = reg2.predict(X_test)
```

```python
plt.scatter(X,Y,label="original data")
plt.plot(X_test,Yt1,color="red",label="depth=2")
plt.plot(X_test,Yt2,color="blue",label="depth=5")
plt.legend()
plt.show()
```
